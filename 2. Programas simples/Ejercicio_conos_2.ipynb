```python
# Declaracion de variables
var_ent_int_Altura = 0.0
var_ent_int_Radio = 0.0
var_proSal_int_Volumen = 0.0 # Variable que almacena
```

```python
# Captura de datos
var_ent_int_Altura = int(input("Altura ;"))
var_ent_int_Radio = int(input("Radio ;"))
```

```python
# proceso
var_proSal_int_Volumen = (3.1416*var_ent_int_Altura*var_ent_int_Radio**2)/3
```
