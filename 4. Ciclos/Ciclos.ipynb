```python
# Declarar variables punto 2
import math

def numero_combinaciones(m,n):
  if m < n:
    return 0
  else:
    comb = math.factorial(m) / (math.factorial(n) * math.factorial(m-n))
    return comb
print(numero_combinaciones(20, 5))
```
