```python
# Tabla de multiplicar de un número
mult=0
tabla= int(input("Que tabla desea : "))
for x in range(1,11):
  mult = tabla * x
  print(str(tabla)+" X "+str(x)+" = "+str(mult))
```

```python
# Tabla de multiplicar de un número
mult=0
tabla= int(input("Que tabla desea : "))
x=1
while x<=11:
  mult = tabla * x
  print(str(tabla)+" X "+str(x)+" = "+str(mult))
  x=x+1
```
