```python
# Declarar variables
num=0
fac=1
# Lecturas
num = int(input("Digite número : "))

# Calcular los procesos
for i in range(1,num+1):
  fac = fac * i

# Imprimir resultados
print("El Factorial es: ", str(fac))

```
