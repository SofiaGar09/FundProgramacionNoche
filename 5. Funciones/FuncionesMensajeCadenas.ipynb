```python
# Desarrollo de la funcion
def buscarPosicionCadena(texto,buscar):
    posicion=texto.find(buscar)
    return posicion

#Declarar variables
# Captura de datos
textoLeer = input("Digite cadena origen : ")
textoLeer = textoLeer.lower()
textoBuscar = input("Digite texto a buscar : ")
textoBuscar = textoBuscar.lower()

# Llamado de la funcion
respuesta = buscarPosicionCadena(textoLeer,textoBuscar)

if respuesta >=0:
  print ("Posicion: ",str(respuesta))
else:
  print ("El texto buscado no existe en la cadena ")
```
