Descripcion del enunciado 3: Construir una funcion que reciba como parametro un entero y retorne la cantidad de digitos 



```python
# Desarrollo de la funcion punto 3
def Contar_Digitos(n):
  cont = 0
  while n > 0:
    cont = cont + 1
    n = n//10
  return cont

# Llamado
print("La cantidad de digitos son:", Contar_Digitos(3547))
```

Declaracion del enunciado 5: Construir una funcion que reciba como parametro un entero y retorne la cantidad de digitos primos. 

```python
# Declaracion variables npunto 5
def CantDigPrimos(n):
  contDigPrimos = 0
  while n > 0:
    UltDig = n % 10
    if UltDig == '2' or UltDig == '3' or UltDig == '5' or UltDig == '7':
      contDigPrimos = contDigPrimos + 1
      n = n / 10
  return(contDigPrimos)
n = int(input("Digite un número: "))
print("La cantidad de numeros primos son: ", CantDigPrimos(n))
```
