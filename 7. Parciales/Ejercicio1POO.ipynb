1.	Crea una clase Complejo que permita trabajar con números complejos (parte real y parte imaginaria). Incluye los siguientes métodos: constructores (por defecto y parametrizado), accedentes, mutadores, suma, resta, multiplicación, división, acumulación y print()

```python
# Suma
class complejo:
  # Metodo Constructor
  def __init__(self, reales, imaginarios):
    self.reales = reales
    self.imaginarios = imaginarios

  def __str__(self):
    return str (self.reales) + "+" + str(self.imaginarios) + "i"

  def sumaComplejo(self, cx):
    rea = self.reales + cx.reales
    ina = self.imaginarios + cx.imaginarios
    c3 = complejo(rea, ina)
    return c3

reales = int(input("Ingrese la parte real: "))
imaginarios = int(input("Ingrese la parte imaginaria: "))

obj_c1 = complejo(reales, imaginarios)
obj_c2 = complejo(5,6)

print(obj_c1)
print(obj_c2)

c4 = obj_c1.sumaComplejo(obj_c2)
print("La suma es: ",c4)

  
```

```python
# Multiplicacion 
class complejo:
  # Metodo Constructor
  def __init__(self, reales, imaginarios):
    self.reales = reales
    self.imaginarios = imaginarios

  def __str__(self):
    return str (self.reales) + "*" + str(self.imaginarios) + "i"

  def multComplejo(self, cx):
    rea = self.reales * cx.reales
    ina = self.imaginarios * cx.imaginarios
    c3 = complejo(rea, ina)
    return c3

reales = int(input("Ingrese la parte real: "))
imaginarios = int(input("Ingrese la parte imaginaria: "))

obj_c1 = complejo(reales, imaginarios)
obj_c2 = complejo(7,8)

print(obj_c1)
print(obj_c2)

c4 = obj_c1.multComplejo(obj_c2)
print(c4)
```

```python
# Division 
class complejo:
  # Metodo Constructor
  def __init__(self, reales, imaginarios):
    self.reales = reales
    self.imaginarios = imaginarios

  def __str__(self):
    return str (self.reales) + "/" + str(self.imaginarios) + "i"

  def divComplejo(self, cx):
    rea = self.reales / cx.reales
    ina = self.imaginarios / cx.imaginarios
    c3 = complejo(rea, ina)
    return c3

reales = int(input("Ingrese la parte real: "))
imaginarios = int(input("Ingrese la parte imaginaria: "))

obj_c1 = complejo(reales, imaginarios)
obj_c2 = complejo(2,3)

print(obj_c1)
print(obj_c2)

c4 = obj_c1.divComplejo(obj_c2)
print(c4)
```

```python
# Resta
class complejo:
  # Metodo Constructor
  def __init__(self, reales, imaginarios):
    self.reales = reales
    self.imaginarios = imaginarios

  def __str__(self):
    return str (self.reales) + "-" + str(self.imaginarios) + "i"

  def restComplejo(self, cx):
    rea = self.reales - cx.reales
    ina = self.imaginarios - cx.imaginarios
    c3 = complejo(rea, ina)
    return c3

reales = int(input("Ingrese la parte real: "))
imaginarios = int(input("Ingrese la parte imaginaria: "))

obj_c1 = complejo(reales, imaginarios)
obj_c2 = complejo(3,4)

print(obj_c1)
print(obj_c2)

c4 = obj_c1.restComplejo(obj_c2)
print(c4)
```
