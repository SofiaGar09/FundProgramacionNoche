```python
class Racional():
  def __init__(self,num,den):
    self.num = num
    self.den = den 

  def __str__(self):
    return str(self.num) + "/" +str(self.den)

frac = Racional (3,4)
print(frac)
```

```python
# Multilicacion 
class Racional():
  def __init__(self,num,den):
    self.num = num
    self.den = den 

  def __str__(self):
    return str(self.num) + "/" +str(self.den)

  def multiplicar(self,otra):
    nuevonum = self.num*otra.num
    nuevoden = self.den*otra.den
    return Racional(nuevonum, nuevoden)

num = int(input("Ingrese un numerador: "))
den = int(input("Ingrese un denominador: "))

obj_frac = Racional(num,den)
print(obj_frac)
obj_frac2 = Racional(2,6)
print(obj_frac2)
resultado = obj_frac.multiplicar(obj_frac2)
print("La multiplicacion es: ", resultado)
```

```python
# Suma 
class Racional():
  def __init__(self,num,den):
    self.num = num
    self.den = den 

  def __str__(self):
    return str(self.num) + "/" +str(self.den)

  def sumar(self,otra):
    nuevonum = self.num+otra.num
    nuevoden = self.den+otra.den
    return Racional(nuevonum, nuevoden)

num = int(input("Ingrese un numerador: "))
den = int(input("Ingrese un denominador: "))

obj_frac = Racional(num,den)
print(obj_frac)
obj_frac2 = Racional(2,6)
print(obj_frac2)
resultado = obj_frac.sumar(obj_frac2)
print("La suma es: ", resultado)
```

```python
# Resta 
class Racional():
  def __init__(self,num,den):
    self.num = num
    self.den = den 

  def __str__(self):
    return str(self.num) + "/" +str(self.den)

  def restar(self,otra):
    nuevonum = self.num-otra.num
    nuevoden = self.den-otra.den
    return Racional(nuevonum, nuevoden)

num = int(input("Ingrese un numerador: "))
den = int(input("Ingrese un denominador: "))

obj_frac = Racional(num,den)
print(obj_frac)
obj_frac2 = Racional(4,7)
print(obj_frac2)
resultado = obj_frac.restar(obj_frac2)
print("La resta es: ", resultado)
```
