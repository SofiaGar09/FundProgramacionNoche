5.	Crea una clase Cuenta (bancaria) con atributos para el número de cuenta (un entero largo), el DNI del cliente (otro entero largo), el saldo actual y el interés anual que se aplica a la cuenta (porcentaje). Define en la clase los siguientes métodos:
 
- Constructor por defecto y constructor con DNI, saldo e interés
- Accedentes y mutadores. Para el número de cuenta no habrá mutador.
- actualizarSaldo(): actualizará el saldo de la cuenta aplicándole el interés diario (interés anual dividido entre 365 aplicado al saldo actual).
-	ingresar(double): permitirá ingresar una cantidad en la cuenta.
-	retirar(double): permitirá sacar una cantidad de la cuenta (si hay saldo).
- Método que nos permita mostrar todos los datos de la cuenta.




```python
class CuentaBancaria():
  numeroCuenta = 100001

  def __init__(self,P_DNI,P_saldo,P_interes):
    self.DNI = P_DNI 
    self.saldoActual = P_saldo
    self.interesAnual = P_interes
    self.numeroCuenta = self.numeroCuenta +1

  # Accedente Numero de Cuenta
  def getNumeroCuenta(self):
    return self.numeroCuenta
  # Accedente DNI
  def getDNI(self):
    return self.DNI  
  # Accedente para saldo
  def getSaldo(self):
    return self.saldoActual
  # Accedente para interes
  def getInteres(self):
    return self.interesAnual

  # Modificador DNI 
  def setDNI(self,DNI):
    self.DNI = DNI
  # Modificador saldo
  def setSaldo(self,saldo):
    self.saldoActual = saldo
  # Modificador interes
  def setInteres(self,interes):
    self.interesAnual = interes
  

  def actualizarSaldo(self):
    interesDiario = self.interesAnual/365 
    saldoActualizado = self.saldoActual + interesDiario
    return saldoActualizado

  def ingresar(self,monto):
    if monto<0:
      print("ERROR: monto no puede ser negativo")
      
    else:
      self.saldoActual+=monto
      print(f"se deposito {monto} en la cuenta {self.DNI}. Nuevo saldo es {self.saldoActual}")

  def retirar(self,monto):
    if monto<0:
      print("ERROR: monto no puede ser negativo")
      
    else:
      self.saldoActual-=monto
      print(f"se retiro {monto} en la cuenta {self.DNI}. Nuevo saldo es {self.saldoActual}")

    

  def toString(self):
    return "Los datos de la cuenta son: " , "Saldo actual: " , self.saldoActual , "DNI: " , self.DNI , "Interes anual: " , self.interesAnual , "Numero de cuenta" , self.numeroCuenta


miCuentaBancaria = CuentaBancaria(30145678, 30000, 0.15)
print(miCuentaBancaria.toString())
print("El saldo actual es: ", miCuentaBancaria.getSaldo())
print("El DNI es: ", miCuentaBancaria.getDNI())
print("El numero de la cuenta es: ", miCuentaBancaria.getNumeroCuenta())
print("El interes es: ", miCuentaBancaria.getInteres())
cuenta = CuentaBancaria(100001, 30000, 0.15)
cuenta.ingresar(300)
cuenta.retirar(5000)
```
