Descripcion del enunciado: Construir una funcion que reciba como parametro un entero y un digito y retorne 1 si dicho digito esta en dicho entero y 0 si no es asi. 

```python
# RECIBIENDO parámetro y  RETORNANADO valor
# Desarrollo de la funcion
def Comparar(n):
  entero = 0
  Dig = 1
  while n>=0:
    if n % 10 <= Dig :
      return(entero)
    else:
      n = n/10
      return(Dig)

# Llamado de la nuncion
n = int(input("Digite numero entero : "))
n = int(input("Digite un digito : "))
print("El digito es :", Comparar(n))

      
```
