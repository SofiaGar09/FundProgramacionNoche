Descripcion del enunciado 2: Construir una funcion que reciba como parametro un entero y retorne sus dos ultimos digitos 

```python
# RECIBIENDO par치metro y  RETORNANADO valor
# Desarrollo de la funcion 
def UltDig2(n):
  UltDig2 = 0
  UltDig2 = n % 100
  return(UltDig2)

# Llamado
n = int(input("Digite numero :"))
print("Sus dos ultimos digitos son :", UltDig2(n))
```

```python
# SIN RECIBIR par치metro y RETORNANDO valor
def UltDig2():
  # Desarrollo de la funcion
  n = int(input("Digite numero :"))
  UltDig2 = 0
  UltDig2 = n % 100
  return(UltDig2)

  
# Llamado
print("Sus dos ultimos digitos son :", UltDig2())
```
