Descripcion del enunciado: Construir una funcion que reciba como parametro un entero y retorne 1 si dicho valor es multiplo de 5. Debera retornar 0 si no es asi

```python
# RECIBIENDO par치metro y  RETORNANADO valor
# Desarrollo de la funcion
def Mult_5(n):
  if n % 5 <= 0:
    return(1)
  else:
    return(0)

#Llamado de la funcion 
n=int(input("Digite numero entero :"))
print("El retorno es : ", Mult_5(n))
```
