```python
# Declaracion de variables
var_ent_int_Vi = 0.0
var_ent_int_Vf = 0.0
var_ent_int_Tiempo = 0.0
var_proSal_int_aceleracion = 0.0 # Variable que almacena
```

```python
# Captura de datos
var_ent_int_Vi = int(input("Velocidad Inicial ;"))
var_ent_int_Vf = int(input("Velocidad Final ;"))
var_ent_int_Tiempo = int(input("Tiempo ;"))
```

Velocidad Inicial: 0

Velocidad Final: 60

```python
# Procesos
var_proSal_int_aceleracion = (var_ent_int_Vf - var_ent_int_Vi)/var_ent_int_Tiempo
```
